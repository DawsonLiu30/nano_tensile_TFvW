Step 01 â€“ Prepare initial nanocrystal structure

This notebook generates the initial Al nanocrystal used for tensile tests.
No grips, no vacancies, no strain are applied at this stage.


```python
A0 = 4.05
SIZE = 3.5
LENGTH_Z = 40.0
GAMMA100 = 1.00
GAMMA110 = 1.06
```

```python
from ase_nanocrystal import build_wulff_nanocrystal
from ase.io import write

atoms = build_wulff_nanocrystal(
    a0=A0,
    size=SIZE,
    length_z=LENGTH_Z,
    gamma100=GAMMA100,
    gamma110=GAMMA110,
)

write("structure_0.xyz", atoms)
write("structure_0.vasp", atoms, vasp5=True, direct=True)

print("Atoms:", len(atoms))
print("Cell:", atoms.get_cell().lengths())
print("Written: structure_0.xyz / structure_0.vasp")
```

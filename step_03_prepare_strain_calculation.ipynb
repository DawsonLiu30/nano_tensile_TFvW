```python
from pathlib import Path
import numpy as np
from ase.io import read, write
import strain_engine

ROOT = Path.cwd()
WORKDIR = ROOT / "cases" / "case_center"

ref_xyz = WORKDIR / "cycle_000_relaxed.xyz"
assert ref_xyz.exists(), f"Missing {ref_xyz}"

bottom_idx = np.load(WORKDIR / "bottom_idx.npy")
top_idx = np.load(WORKDIR / "top_idx.npy")

atoms = read(str(ref_xyz))

print("N_atoms =", len(atoms))
print("Bottom atoms =", len(bottom_idx))
print("Top atoms    =", len(top_idx))
```

```python
strain_axis = 2
strain_step = 0.01
cycle = 1

print("Strain axis =", strain_axis)
print("Strain step =", strain_step)
```

```python
atoms_stretched, info = strain_engine.apply_strain(
    atoms,
    strain_step,
    bottom_idx,
    top_idx,
    axis=strain_axis,
    debug=True,
    return_info=True,
)

out_xyz = WORKDIR / f"cycle_{cycle:03d}_stretched.xyz"
write(str(out_xyz), atoms_stretched)

print("Saved:", out_xyz)
print("Gauge_L_before_A =", f"{info['L']:.6f}")
print("dL_A             =", f"{info['dL']:.6f}")
```

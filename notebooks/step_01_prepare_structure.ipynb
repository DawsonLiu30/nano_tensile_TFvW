# Step 01 - Prepare Structure

Create a nanocrystal structure and write it to `cases/<CASE_NAME>/inputs`.

```python
from pathlib import Path
import sys


def find_project_root(start: Path) -> Path:
    cur = start.resolve()
    for candidate in [cur, *cur.parents]:
        has_root = (candidate / "main.py").exists()
        has_pp = (candidate / "al.gga.recpot").exists() or (candidate / "al.gga.psp").exists()
        if has_root and has_pp:
            return candidate
    raise RuntimeError("Project root not found (expected main.py and Al pseudopotential file).")


ROOT = find_project_root(Path.cwd())
if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))

from app.aluminum_defaults import AL_FCC_A0_TFVW_ANG

CASE_NAME = "nc_large_vac01"

# Geometry controls
A0 = AL_FCC_A0_TFVW_ANG
SIZE = 6.5
LENGTH_Z = 60.0
VACUUM = 5.0
GAMMA100 = 1.00
GAMMA110 = 1.06

CASE_DIR = ROOT / "cases" / CASE_NAME
INPUTS_DIR = CASE_DIR / "inputs"
INPUTS_DIR.mkdir(parents=True, exist_ok=True)

print("ROOT:", ROOT)
print("CASE_DIR:", CASE_DIR)
print("INPUTS_DIR:", INPUTS_DIR)
print("A0 (TFvW bulk equilibrium):", A0)
```

```python
from app.ase_nanocrystal import build_wulff_nanocrystal
from ase.io import write

atoms = build_wulff_nanocrystal(
    a0=A0,
    size=SIZE,
    length_z=LENGTH_Z,
    vacuum=VACUUM,
    gamma100=GAMMA100,
    gamma110=GAMMA110,
)

write(str(INPUTS_DIR / "raw_structure.xyz"), atoms)
write(str(INPUTS_DIR / "raw_structure.vasp"), atoms, vasp5=True, direct=True)

print("Atoms:", len(atoms))
print("Cell (A):", atoms.get_cell().lengths())
print("Written:", INPUTS_DIR / "raw_structure.xyz")
print("Written:", INPUTS_DIR / "raw_structure.vasp")
```
